## limiting to chicago and columns I've determined interesting

```python
cities = general2020["recipient_city"].value_counts()
cities["CHICAGO"], cities["Chicago"], cities["chicago"] #different ways chicago could be recorded
```

```python
chicago2020 = general2020.query('recipient_city == "Chicago" or recipient_city ==  "CHICAGO" or recipient_city == "chicago"')
chicago2020 = chicago2020[['Recipient_Zip_Code', 'Physician_Specialty',\
                          'Total_Amount_of_payment_usdollars', 'date_of_payment','number_of_payments_included_in_total_amount',\
                          'form_of_payment_or_transfer_of_value','nature_of_payment_or_transfer_of_value',]]
```

```python
chicago2015 = general2015.query('Recipient_City == "Chicago" or Recipient_City ==  "CHICAGO" or Recipient_City == "chicago"')
chicago2015 = chicago2015[['Recipient_Zip_Code', 'Physician_Specialty',\
                          'Total_Amount_of_Payment_USDollars', 'Date_of_Payment','Number_of_Payments_Included_in_Total_Amount',\
                          'Form_of_Payment_or_Transfer_of_Value','Nature_of_Payment_or_Transfer_of_Value',]]
```
